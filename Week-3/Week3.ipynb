```python
import pandas as pd
df= pd.read_csv('netflix.csv')
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title'])
df['description'] = df['description'].fillna('No description available')
df['age_certification'] = df['age_certification'].fillna('Not Rated')
df['seasons'] = df['seasons'].fillna(0)
df['type'] = df['type'].str.upper()
df['type'].value_counts()
df['release_year'] = df['release_year'].astype(int)
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['content_age'] = 2025 - df['release_year']
```

```python
df['release_decade'] = (df['release_year'] // 10) * 10
df['release_decade'] = df['release_decade'].astype(str) + 's'
```

```python
year_type_counts = (
    df.groupby(['release_year', 'type'])
      .size()
      .unstack(fill_value=0)
)
year_type_counts
```

```python
year_type_counts['movie_tv_ratio'] = (
    year_type_counts['MOVIE'] /
    year_type_counts['SHOW'].replace(0, 1)
)
df = df.merge(
    year_type_counts['movie_tv_ratio'],
    on='release_year',
    how='left'
)
```

```python
df.to_csv("netflix_final_features.csv", index=False)
```
